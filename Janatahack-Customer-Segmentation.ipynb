```
import numpy as np
Gender = {
    'Male':0,
    'Female':1,
    np.nan:10
}
Ever_Married = {
    'Yes':0,
    'No':1,
    np.nan:10
}
Graduated = {
    'Yes':0,
    'No':1,
    np.nan:10
}
Profession = {
    'Healthcare':0,
    'Engineer':1,
    'Lawyer':2,
    'Entertainment':3,
    'Artist':4,
    'Executive':5,
    'Doctor':6,
    'Homemaker':7,
    'Marketing':8,
    np.nan:10
}
Spending_Score = {
    'Low':0,
    'Average':1,
    'High':2,
    np.nan:10
}
Var_1 = {
    'Cat_1':0,
    'Cat_2':1,
    'Cat_3':2,
    'Cat_4':3,
    'Cat_5':4,
    'Cat_6':5,
    'Cat_7':6,
    np.nan:10
}
to_change = [Gender, Ever_Married, Graduated, Profession, Spending_Score, Var_1] 
for i in to_change:
  train = train.replace(i)
  test = test.replace(i)
```

```
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```
X = train
X.shape
```

```
X.head()
```

```
y = X['Segmentation']
X = X.drop(['ID', 'Segmentation'], axis=1)
```

```
X.head()
```

```
y.head()
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```
from sklearn.ensemble import RandomForestClassifier
clf_rf = RandomForestClassifier(
    max_depth=2, 
    random_state=0,
    n_estimators=500,
    
)

clf_rf.fit(X_train, y_train)
y_pred = clf_rf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```
from sklearn.ensemble import RandomForestClassifier
clf_svm = RandomForestClassifier(
    n_estimators=1000
)

clf_svm.fit(X_train, y_train)
y_pred = clf_svm.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
